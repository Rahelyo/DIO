```python
import speech_recognition as sr
import pyttsx3
import webbrowser
import wikipedia


# Configuração do módulo 1: Text to Speech (Texto para Áudio)
def falar(texto):
    engine = pyttsx3.init()
    engine.say(texto)
    engine.runAndWait()


# Configuração do módulo 2: Speech to Text (Fala para Texto)
def ouvir_comando():
    reconhecedor = sr.Recognizer()

    with sr.Microphone() as source:
        print("Ouvindo...")
        reconhecedor.adjust_for_ambient_noise(source)
        audio = reconhecedor.listen(source)

    try:
        comando = reconhecedor.recognize_google(audio, language='pt-BR')
        print(f"Você disse: {comando}")
        return comando.lower()
    except Exception as e:
        print("Não entendi, pode repetir?")
        return ""


# Módulo de Automação de Comandos
def executar_assistente():
    falar("Olá! Como posso te ajudar hoje?")

    while True:
        comando = ouvir_comando()

        if 'abrir wikipedia' in comando:
            falar("O que você deseja pesquisar?")
            pesquisa = ouvir_comando()
            resultado = wikipedia.summary(pesquisa, sentences=2)
            falar(f"Segundo a Wikipedia: {resultado}")

        elif 'abrir youtube' in comando:
            falar("Abrindo o YouTube.")
            webbrowser.open("https://www.youtube.com")

        elif 'farmácia' in comando:
            falar("Localizando a farmácia mais próxima no Google Maps.")
            webbrowser.open("https://www.google.com/maps/search/farmacia+proxima")

        elif 'parar' in comando or 'sair' in comando:
            falar("Até logo!")
            break


if __name__ == "__main__":
    executar_assistente()
```
