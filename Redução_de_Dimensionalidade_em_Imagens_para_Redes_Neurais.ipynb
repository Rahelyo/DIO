```python
import cv2
import matplotlib.pyplot as plt

# 1. Carregar a imagem colorida
# Certifique-se de que o arquivo 'imagem.jpg' esteja na mesma pasta do script
img_original = cv2.imread('FotoAjuste.jpg')
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB) # Converter de BGR para RGB para exibição correta

# 2. Transformar para Níveis de Cinza (0 a 255)
img_gray = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)

# 3. Binarização (Preto e Branco - 0 ou 255)
# O valor 127 é o limiar (threshold). Pixels acima de 127 tornam-se 255 (branco), abaixo tornam-se 0 (preto).
resultado, img_binaria = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

# --- Exibição dos Resultados ---
titulos = ['Imagem Original', 'Imagem em Cinza', 'Imagem Binária']
imagens = [img_rgb, img_gray, img_binaria]

plt.figure(figsize=(15, 5))

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(imagens[i], cmap='gray' if i > 0 else None)
    plt.title(titulos[i])
    plt.axis('off')

plt.show()
```
